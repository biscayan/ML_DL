```python
import torch
import torch.optim as optim

#dataset
x_data=[[1,2],
        [2,3],
        [3,1],
        [4,3],
        [5,3],
        [6,2]]

y_data=[[0],
        [0],
        [0],
        [1],
        [1],
        [1]]

#training data
x_train=torch.FloatTensor(x_data) #6x2
y_train=torch.FloatTensor(y_data) #6x1

#training variable
W=torch.zeros((2,1),requires_grad=True)
b=torch.zeros(1,requires_grad=True)



optimizer=optim.SGD([W,b],lr=0.1)

epochs=2000

for epoch in range(epochs+1):
    hypothesis=1/(1+torch.exp(-(torch.matmul(x_train,W)+b)))

    losses=-(y_train*torch.log(hypothesis)+(1-y_train)*torch.log(1-hypothesis))
    cost=losses.mean()
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch %50==0:
        print('Epoch: {:4d} | Cost: {:.6f}'.format(epoch,cost.item()))


```

```python
import torch
import torch.optim as optim
import torch.nn.functional as F

#dataset
x_data=[[1,2],
        [2,3],
        [3,1],
        [4,3],
        [5,3],
        [6,2]]

y_data=[[0],
        [0],
        [0],
        [1],
        [1],
        [1]]

#training data
x_train=torch.FloatTensor(x_data) #6x2
y_train=torch.FloatTensor(y_data) #6x1

W=torch.zeros((2,1),requires_grad=True)
b=torch.zeros(1,requires_grad=True)

#training variable

optimizer=optim.SGD([W,b],lr=0.1)

epochs=2000

for epoch in range(epochs+1):
    hypothesis=torch.sigmoid(torch.matmul(x_train,W)+b)
    cost=F.binary_cross_entropy(hypothesis,y_train)
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch %50==0:
        print('Epoch: {:4d} | Cost: {:.6f}'.format(epoch,cost.item()))
```

```python
import torch
import torch.optim as optim
import torch.nn.functional as F
import torch.nn as nn

class BinaryClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(self.linear(x))
    
#dataset
x_data=[[1,2],
        [2,3],
        [3,1],
        [4,3],
        [5,3],
        [6,2]]

y_data=[[0],
        [0],
        [0],
        [1],
        [1],
        [1]]

#training data
x_train=torch.FloatTensor(x_data) #6x2
y_train=torch.FloatTensor(y_data) #6x1
  
model=BinaryClassifier()
optimizer=optim.SGD(model.parameters(),lr=0.1)
epochs=1000

#training
for epoch in range(epochs+1):
    hypothesis=model(x_train)
    cost=F.binary_cross_entropy(hypothesis,y_train)
    
    optimizer.zero_grad
    cost.backward()
    optimizer.step()
    
    if epoch % 50 ==0:
        print('Epoch: {:4d} | Cost: {:.6f}'.format(epoch,cost.item()))

#accuracy test
hypothesis=torch.sigmoid(torch.matmul(x_train,W)+b)
prediction=hypothesis>=torch.FloatTensor([0.5])
print(hypothesis)
print(prediction.float())
print(y_train)
correct_prediction = prediction.float() == y_train
print(correct_prediction)
accuracy = correct_prediction.sum().item() / len(correct_prediction)
print('The model has an accuracy of {:.2f}% for the training set.'.format(accuracy * 100))
```
